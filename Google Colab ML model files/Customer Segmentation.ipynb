```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
customer_data=pd.read_csv('/content/Mall_Customers.csv')
customer_data.head()
```

```python
customer_data.shape
```

```python
customer_data.info()
```

```python
x=customer_data.iloc[:,[3,4]].values
print(x)
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,n_init=10,random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title('The Elbow point Graph')
plt.xlabel('No.of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,n_init='auto',random_state=0)
y=kmeans.fit_predict(x)
```

```python
print(y)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x[y==0,0],x[y==0,1],s=50,c='green',label='C1')
plt.scatter(x[y==1,0],x[y==1,1],s=50,c='blue',label='C2')
plt.scatter(x[y==2,0],x[y==2,1],s=50,c='pink',label='C3')
plt.scatter(x[y==3,0],x[y==3,1],s=50,c='yellow',label='C4')
plt.scatter(x[y==4,0],x[y==4,1],s=50,c='orange',label='C5')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='black',label='Centroid')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
